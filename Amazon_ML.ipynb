```python
import cv2
import numpy as np
import pandas as pd
import easyocr
import requests
from PIL import Image
from io import BytesIO

# Initialize EasyOCR reader
reader = easyocr.Reader(['en'])

# Load the CSV file
data = pd.read_csv("train.csv")
new_data = data.head(100)

# Function to preprocess the image
def preprocess_image(image_link):
    # Download the image from the link
    response = requests.get(image_link)
    img = Image.open(BytesIO(response.content))

    # Convert the image to a NumPy array for OpenCV processing
    img_array = np.array(img)

    # Convert to grayscale
    gray_image = cv2.cvtColor(img_array, cv2.COLOR_BGR2GRAY)

    # Apply binary thresholding
    _, threshold_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)

    # Optionally apply noise reduction (median blur)
    preprocessed_image = cv2.medianBlur(threshold_image, 3)

    # Convert back to PIL Image (since EasyOCR works with PIL format)
    return Image.fromarray(preprocessed_image)

# Function to extract text from a single image link with preprocessing
def extract_text_from_image(image_link):
    preprocessed_img = preprocess_image(image_link)
    # Extract text using EasyOCR
    result = reader.readtext(np.array(preprocessed_img), detail=0)
    return ' '.join(result)  # Join the text blocks into a single string

# Loop through each image link and extract text
extracted_texts = []

for image_link in new_data["image_link"]:
    try:
        extracted_text = extract_text_from_image(image_link)
        extracted_texts.append(extracted_text)
    except Exception as e:
        extracted_texts.append("")  # Append an empty string in case of an error
        print(f"Error processing image {image_link}: {e}")

# Add the extracted text as a new column in the DataFrame
new_data['extracted_text'] = extracted_texts

# Check the DataFrame
print(new_data.head())
```

```python
import spacy
import pandas as pd
from spacy.training import Example
from spacy.util import minibatch
import random

# Create empty list to store training data
TRAIN_DATA = []

# Function to find the start and end positions of entity in text
def find_entity_position(text, entity_value):
    if pd.notna(entity_value):  # Check if the entity value is not NaN
        start = text.find(entity_value)
        if start != -1:
            end = start + len(entity_value)
            return (start, end)
    return None

# Iterate over each row and extract entities
for index, row in new_data.iterrows():
    text = row['extracted_text']  # The sentence or description

    # Initialize an empty list of entities
    entities = []

    # Extract entities without overlaps
    for entity_name in ['WEIGHT', 'VOLT', 'DIMENSION', 'VOLUME']:
        entity_value = row['entity_name']
        if pd.notna(entity_value):
            pos = find_entity_position(text, entity_value)
            if pos and not any(pos[0] < e[1] and pos[1] > e[0] for e in entities):
                entities.append((pos[0], pos[1], entity_name))

    # If entities were found, add to training data
    if entities:
        TRAIN_DATA.append((text, {"entities": entities}))

# Load or create a blank spaCy model
nlp = spacy.blank("en")  # Start with a blank English model

# Add the NER component to the pipeline if it's not already present
if "ner" not in nlp.pipe_names:
    ner = nlp.add_pipe("ner")

# Add custom labels to the NER
labels = ["WEIGHT", "DIMENSION", "VOLUME", "VOLT"]
for label in labels:
    ner.add_label(label)

# Disable other pipeline components and train only NER
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != "ner"]
with nlp.disable_pipes(*other_pipes):
    optimizer = nlp.begin_training()

    # Training loop
    for iteration in range(30):  # You can adjust the number of iterations
        random.shuffle(TRAIN_DATA)
        losses = {}
        batches = minibatch(TRAIN_DATA, size=8)
        for batch in batches:
            for text, annotations in batch:
                doc = nlp.make_doc(text)
                example = Example.from_dict(doc, annotations)
                nlp.update([example], drop=0.35, losses=losses)
        print(f"Iteration {iteration} Losses: {losses}")

# Save the trained model
nlp.to_disk("./new_custom_ner_model")
print("NER model saved!")
```

```python
# Load the saved model
nlp = spacy.load("./new_custom_ner_model")

# Test the model on a new sentence
doc = nlp("The shipment weighs 300 grams and its dimensions are 20x30 cm.")

# Print the detected entities and their labels
for ent in doc.ents:
    print(f"Entity: {ent.text}, Label: {ent.label_}")
```
